```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Data Prepration

```python
data_s = pd.read_csv('Warehouse_and_Retail_sales.csv')
```

```python
data_s.info()
```

```python
data_s.describe(include='all')
```

```python
data_s.isna().sum()
```

```python
data_s.dropna(subset=['ITEM TYPE', 'RETAIL SALES'], inplace=True)
```

Exploratory Data Analysis

```python
#Visualization of each year's total sales
yearly_sales = data_s.groupby('YEAR')[['RETAIL SALES', 'WAREHOUSE SALES']].sum()
yearly_sales.plot(kind='bar', stacked=True)
plt.title('Yearly Total Sales (Retail and Warehouse)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Top 5 suppliers
supplier_sales = data_s.groupby('SUPPLIER')[['RETAIL SALES', 'WAREHOUSE SALES']].sum()
supplier_sales['TOTAL SALES'] = supplier_sales['RETAIL SALES'] + supplier_sales['WAREHOUSE SALES']
top_suppliers = supplier_sales.sort_values(by='TOTAL SALES', ascending=False).head(5)
print(top_suppliers)
```

```python
#Visualization of top 5 suppliers
plt.figure(figsize=(10, 6))
plt.bar(top_suppliers.index, top_suppliers['TOTAL SALES'], color='skyblue')
plt.title('Top 5 Suppliers Based on Total Sales')
plt.xlabel('Supplier')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Best selling items
item_sales = data_s.groupby('ITEM CODE')[['RETAIL SALES', 'WAREHOUSE SALES']].sum()
item_sales['TOTAL SALES'] = item_sales['RETAIL SALES'] + item_sales['WAREHOUSE SALES']
top_items = item_sales.sort_values(by='TOTAL SALES', ascending=False).head(10)
top_items = top_items.merge(data_s[['ITEM CODE', 'ITEM DESCRIPTION', 'ITEM TYPE']], on='ITEM CODE')
print(top_items[['ITEM CODE', 'ITEM DESCRIPTION', 'ITEM TYPE', 'TOTAL SALES']])
```

```python
#Visualization of 10 Best selling items
plt.figure(figsize=(12, 6))
plt.bar(top_items['ITEM DESCRIPTION'], top_items['TOTAL SALES'], color='skyblue')
plt.title('Top 10 Best-Selling Items Based on Total Sales')
plt.xlabel('Item Description')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

Business Analysis

```python
#Monthly average retail and warehouse sales year wise
monthly_sales = data_s.groupby(['YEAR', 'MONTH'])[['RETAIL SALES', 'WAREHOUSE SALES']].mean()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='MONTH', y='RETAIL SALES', hue='YEAR', marker='o', markersize=8)
plt.title('Monthly Average Retail Sales Over Years')
plt.xlabel('Month')
plt.ylabel('Average Retail Sales')
plt.legend(title='Year')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='MONTH', y='WAREHOUSE SALES', hue='YEAR', marker='o', markersize=8)
plt.title('Monthly Average Warehouse Sales Over Years')
plt.xlabel('Month')
plt.ylabel('Average Warehouse Sales')
plt.legend(title='Year')
plt.show()
```

```python
#Visualization of total sales of each item
item_type_sales = data_s.groupby('ITEM TYPE')[['RETAIL SALES', 'WAREHOUSE SALES']].sum()
item_type_sales['TOTAL SALES'] = item_type_sales['RETAIL SALES'] + item_type_sales['WAREHOUSE SALES']
item_type_sales = item_type_sales.sort_values(by='TOTAL SALES', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(item_type_sales.index, item_type_sales['TOTAL SALES'], color='skyblue')
plt.title('Total Sales by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
#Item type with the highest total sales
most_sold_item_type = item_type_sales['TOTAL SALES'].idxmax()
total_sales_for_most_sold_item_type = item_type_sales['TOTAL SALES'].max()
print(f"The item type contributing the most to overall sales is '{most_sold_item_type}' with a total sales of ${total_sales_for_most_sold_item_type:,.2f}")
```
